```python
import requests
import json
from datetime import datetime, timedelta

#Konfigurasi Koordinat Bandar Lampung
LATITUDE = -5.4292
LONGITUDE = 105.2618
KOTA = "Bandar Lampung"

#Rentang tanggal: 1 minggu ke belakang
hari_ini = datetime.today()
tanggal_mulai = (hari_ini - timedelta(days=6)).strftime("%Y-%m-%d")
tanggal_akhir = hari_ini.strftime("%Y-%m-%d")

url = "https://api.open-meteo.com/v1/forecast"

params = {
    "latitude": LATITUDE,
    "longitude": LONGITUDE,
    "daily": ["temperature_2m_max", "temperature_2m_min", "precipitation_sum", "windspeed_10m_max"],
    "timezone": "Asia/Jakarta",
    "start_date": tanggal_mulai,
    "end_date": tanggal_akhir
}

response = requests.get(url, params=params)
data = response.json()

print(f"Status API: {response.status_code}")
print(f"Mengambil data cuaca {KOTA} dari {tanggal_mulai} s/d {tanggal_akhir}\n")

#Simpan ke file
nama_file = "cuaca_bandar_lampung.txt"

tanggal_list   = data["daily"]["time"]
suhu_max_list  = data["daily"]["temperature_2m_max"]
suhu_min_list  = data["daily"]["temperature_2m_min"]
hujan_list     = data["daily"]["precipitation_sum"]
angin_list     = data["daily"]["windspeed_10m_max"]

with open(nama_file, "a", encoding="utf-8") as f:  #mode append
    f.write(f"\n{'='*50}\n")
    f.write(f"Data Cuaca {KOTA}\n")
    f.write(f"Diambil pada: {hari_ini.strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write(f"{'='*50}\n")

    #LOOP untuk setiap hari (Tugas 3)
    for i in range(len(tanggal_list)):
        tanggal  = tanggal_list[i]
        suhu_max = suhu_max_list[i]
        suhu_min = suhu_min_list[i]
        hujan    = hujan_list[i]
        angin    = angin_list[i]

        #KONDISI: hanya simpan data yang layak/menarik (Tugas 4)
        if suhu_max is not None and suhu_min is not None:

            # Tentukan kategori cuaca
            if hujan > 10:
                kategori = "HUJAN LEBAT"
            elif hujan > 0:
                kategori = "Hujan Ringan"
            else:
                kategori = "Cerah / Tidak Hujan"

            #Kondisi tambahan: tandai hari ekstrem
            peringatan = ""
            if suhu_max > 35:
                peringatan = " ⚠️ SUHU SANGAT PANAS"
            if angin > 30:
                peringatan += " ⚠️ ANGIN KENCANG"

            baris = (
                f"Tanggal  : {tanggal}\n"
                f"Suhu     : {suhu_min}°C - {suhu_max}°C\n"
                f"Hujan    : {hujan} mm\n"
                f"Angin    : {angin} km/h\n"
                f"Cuaca    : {kategori}{peringatan}\n"
                f"{'-'*40}\n"
            )

            f.write(baris)
            print(baris, end="")  # tampilkan juga di output

print(f"\n✅ Data berhasil disimpan ke file: {nama_file}")
```
