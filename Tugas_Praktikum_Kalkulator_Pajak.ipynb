```python
from diskon import hitung_diskon
from pajak import hitung_ppn

#data transaksi
harga_awal = 1000000
diskon_persen = 10

potongan_diskon = hitung_diskon(harga_awal, diskon_persen)
harga_setelah_diskon = harga_awal - potongan_diskon
besaran_ppn = hitung_ppn(harga_setelah_diskon)
total_bayar = harga_setelah_diskon + besaran_ppn

print(f"Harga Awal: Rp", (harga_awal))
print(f"Besaran Potongan Diskon: Rp", (potongan_diskon))
print(f"Harga Setelah Diskon: Rp", (harga_setelah_diskon))
print(f"Besaran PPN: Rp", (besaran_ppn))
print(f"Total Yang Harus Dibayar: Rp", (total_bayar))
```
