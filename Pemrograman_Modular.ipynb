**PEMROGRAMAN MODULAR**

Proses memecah program besar menjadi bagian kecil yang terpisah
pendektan ini adalah akar dari modul

**Pros**

simplicity: kode program menjadi jauh lebih sederhana dan tdk menumpuk
maintainabillity: kode program lebih mudah dirawat , diperbaiki, dan dikembangkan
reusability: potongan kode yg bisa digunakan berulang kali dari berbagai tempat tanpa harus menulis ulang

**Komponen**
Fungsi: blok kode bernama untuk menyelesaikan satu tugas spesifik
modul: file berisi kumpulan skrip/kode phyton
paket: direktori/folder berisi kumpulan modul phyton

**Modul**
modul adalah sebuah file bereksistensi .py yang berisi skrip phyton
nama modul diambil dr nama file tersebut
modul bisa berisi fungsi, variabel, maupun class

Pemrograman modular memecah program menjadi bagian bermakna
modul adalah file .py sedangkan paket adalah direktori berisi modul dan __init__.py.
tujuannya adalah membuat kode lebih sederhana, mudah dikelola, dan dpt digunakan ulang (reusable)
