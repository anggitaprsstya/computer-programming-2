JSON atau Javascript Object Notation adalah format teks standar yang digunakan untuk mengirimkan dan menyimpan data terstruktur.

Kenapa API penting? hampir semua aplikasi digital modern (dari medsos hingga ecommerce) bergantung pada API untuk berfungsi. Contohnya seperti bot telegram dan whatsapp, chatbot ai, aplikasi mobile, website dan dashboard.

API adalah jembatan komunikasi, program kita bisa berbicara dengan sistem lain di seluruh dunia melalui API. Seperti mengambil data cuaca, kutipan, prediksi, bahkan kecerdasan buatan.

Alur program API yaitu request > response > JSON > output

```python
#Contoh data JSON
{
    "name": "Budi",
    "age": 25,
    "city": "Jakarta"
}
```

```python
#PRAKTIK 1: Mengirim Request ke API
import requests

url = "https://api.agify.io?name=budi"
response = requests.get(url)

print(response)
print(response.text)
```

```python
#PRAKTIK 2: Menggunakan Data JSON

data = response.json()

print(data)
print(data["name"])
print(data["age"])
```

```python
#PRAKTIK 3: Input dari Pengguna

import requests

name = input("Masukkan Nama: ")
url = f"https://api.agify.io?name={name}"

response = requests.get(url)
data = response.json()

print(f"Halo, {data['name']}!")
print(f"Perkiraan umur: {data['age']} tahun")
```

```python
#PRAKTIK 4: API Quotes - Kutipan Inspiratif

import requests

url= "https://quotes.liupurnomo.com/api/quotes/random"
data = requests.get(url).json()

print(data["data"]["text"])
print("-", data["data"]["author"])
```

```python
#PRAKTIK 5: Looping dengan API

import requests
for i in range(9):
  url = "https://quotes.liupurnomo.com/api/quotes?category=motivasi&page=1&limit=10"
data = requests.get(url).json()

print(f"--- Quote ke-{i+1} ---")
print(data["data"][i+1]["text"])
print(f"-{data["data"][i+1]['author']}")
print()
```
